```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
import scipy.stats as st
from scipy.stats import linregress
import seaborn as sns

import folium
from folium.plugins import HeatMap
```

```python
cali_fire = "Resources/California_Fire_Incidents.csv"

cali_fire_df = pd.read_csv(cali_fire)
cali_fire_df.head(10)
```

```python
# check coordinates; there are quite a few unrealistic ones
plt.scatter(cali_fire_df.Longitude, cali_fire_df.Latitude)
plt.grid()
plt.show()
```

```python
# remove those rows having unrealistic coordinates
cali_fire_df_select = cali_fire_df[cali_fire_df.Longitude<-115]
cali_fire_df_select = cali_fire_df_select[(cali_fire_df_select.Latitude<44) & (cali_fire_df_select.Latitude > 30)]
plt.scatter(cali_fire_df_select.Longitude, cali_fire_df_select.Latitude)
plt.grid()
plt.show()
```

```python
# ok still one outlier in Nevada, let's get rid of that too
outlier = cali_fire_df_select[cali_fire_df_select.CanonicalUrl=='/incidents/2013/8/6/tram-fire/']
print('Outlier Lon/Lat:', outlier.Longitude, outlier.Latitude)
```

```python
cali_fire_df_select = cali_fire_df_select[cali_fire_df_select.CanonicalUrl!='/incidents/2013/8/6/tram-fire/']
plt.scatter(cali_fire_df_select.Longitude, cali_fire_df_select.Latitude)
plt.grid()
plt.show()
```

```python
# interactive map
zoom_factor = 5 # inital map size
radius_scaling = 50 # scaling of bubbles

my_map_1 = folium.Map(location=[36,-120], zoom_start=zoom_factor)

for i in range(0,cali_fire_df_select.shape[0]):
   folium.Circle(
      location=[cali_fire_df_select.iloc[i]['Latitude'], cali_fire_df_select.iloc[i]['Longitude']],
      radius=np.sqrt(cali_fire_df_select.iloc[i]['AcresBurned'])*radius_scaling,
      color='red',
      popup='CanonicalUrl:' + cali_fire_df_select.iloc[i]['CanonicalUrl'] + ' - Year:' + str(int(cali_fire_df_select.iloc[i]['ArchiveYear'])) + ' - Acres Burned:' 
      + str(cali_fire_df_select.iloc[i]['AcresBurned']),
      fill=True,
      fill_color='red'
   ).add_to(my_map_1)

my_map_1 # display
```

```python
# use heatmap
my_map_2 = folium.Map(location=[36,-120], zoom_start=zoom_factor)
HeatMap(data=cali_fire_df_select[['Latitude', 'Longitude']], radius=10).add_to(my_map_2)

my_map_2 # display
```
